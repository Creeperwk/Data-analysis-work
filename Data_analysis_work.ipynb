```python
import pandas as pd
import matplotlib.pyplot as plt
whiskies = pd.read_csv('https://raw.githubusercontent.com/UofGAnalyticsData/DPIP/main/whiskies.csv') #import the data

plt.plot('Latitude','Longitude','.',data=whiskies) #Scatter plot
plt.xlabel('Latitude')
plt.ylabel('Longitude')
lat_range = max(whiskies['Latitude'])-min(whiskies['Latitude']) #Setting the range
long_range = max(whiskies['Longitude'])-min(whiskies['Longitude'])
plt.figure(figsize=(lat_range/50000,long_range/50000))



```

```python
colours = whiskies["Region"].map({"Highland":"blue","Speyside":"red","Islay":"purple","Lowland":"teal","Campbeltown":"yellow"}) #Classify different groups by colours
plt.scatter("Latitude","Longitude", c=colours , data=whiskies )
plt.show()

```

```python
counts = whiskies.groupby("Region").apply(lambda x:x.shape[0])
plt.pie(counts,colors=["yellow", "blue", "purple","teal","red"],
        labels=["Campbeltown", "Highland","Islay","Lowland","Speyside"] ) #Pie chart of different regions
```

```python
from scipy import stats
import numpy as np

# Create three variables with 1000 samples each and normally distributed with mean 1,3,3 respectively
n1 = stats.norm(1,1)
n3 = stats.norm(3,1)
x1 = n1.rvs(1000)
x2 = n3.rvs(1000)
x3 = n3.rvs(1000)
```

```python
import pandas as pd
x = pd.DataFrame({"x1":x1,"x2":x2,"x3":x3},columns=["x1","x2","x3"]) #Create a dataframe containing three variables
```

```python
x.describe() # Summary of the data
```

```python
# Two sample t-test between pairs of variables
print(stats.ttest_ind(x1,x2))
print(stats.ttest_ind(x1,x3))
print(stats.ttest_ind(x2,x3))
```

```python
# Kernel density plots
import matplotlib.pyplot as plt
x0= np.linspace(-4,8,1000)
plt.plot(x0,stats.gaussian_kde(x1)(x0),label="Kernel density rate for x1")
plt.plot(x0,stats.gaussian_kde(x2)(x0),label="Kernel density rate for x2")
plt.plot(x0,stats.gaussian_kde(x3)(x0),label="Kernel density rate for x3")
plt.legend(loc='center left')
plt.show()
```

```python
# Import the dataset
from sklearn.datasets import load_breast_cancer
dataset = load_breast_cancer()
X= dataset.data
y= dataset.target

# Means and sd for each covariates
print(X.mean(axis=0))
print(X.std(axis=0))

```

```python
# Scale the data
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
X_sca = scalar.fit_transform(X)
print(X_sca)
```

```python
# Split the scaled data into training and test datasets
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_sca,y, random_state=1, test_size=0.2)

```

```python
# Fit a logistic model on the training data
from sklearn.linear_model import LogisticRegression

log_regression = LogisticRegression()
log_regression.fit(X_train,y_train)

y_train_pred = log_regression.predict(X_train)
y_test_pred = log_regression.predict(X_test)
```

```python
# Accuracy and F1 score of the training and test data
from sklearn import metrics
print("Test F1 score:",metrics.f1_score(y_test,y_test_pred))
print("Train F1 score:",metrics.f1_score(y_train,y_train_pred))
print("Test Accuracy:",metrics.accuracy_score(y_test,y_test_pred))
print("Train Accuracy:",metrics.accuracy_score(y_train,y_train_pred))


```

```python
# Using 50% of the best covariates and visualise it with a bar plot
from sklearn.feature_selection import GenericUnivariateSelect,mutual_info_classif
transformer = GenericUnivariateSelect(mutual_info_classif,mode='percentile',param=50)
X_fs = transformer.fit(X_train,y_train)
print(X_fs.get_support)
plt.bar(range(30),X_fs.scores_)
```

```python
#Using the selected model and predict again
X_train_updated = transformer.fit_transform(X_train,y_train)
X_test_updated = transformer.fit_transform(X_test,y_test)
log_regression = LogisticRegression()
log_regression.fit(X_train_updated,y_train)
log_regression.fit(X_test_updated,y_test)

y_train_up_pred = log_regression.predict(X_train_updated)
y_test_up_pred = log_regression.predict(X_test_updated)

print("Test F1 score updated:",metrics.f1_score(y_test,y_test_up_pred))
print("Train F1 score updated:",metrics.f1_score(y_train,y_train_up_pred))
print("Test Accuracy updated:",metrics.accuracy_score(y_test,y_test_up_pred))
print("Train Accuracy updated:",metrics.accuracy_score(y_train,y_train_up_pred))
```
